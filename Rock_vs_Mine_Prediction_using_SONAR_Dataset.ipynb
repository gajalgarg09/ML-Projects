Importing the libraries


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Data Preprocessing

```python
#loading the dataset to pandas dataframe
sonar_data = pd.read_csv('SONAR_Dataset.csv' , header = None)
```

```python
sonar_data.head()
```

```python
# number of rows and columns
sonar_data.shape
```

```python
# used to find the statistical measures of the data
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
# 60 -> column value
```

M -> Mine <br>
R -> Rock

```python
sonar_data.groupby(60).mean
```

```python
# separating data and labels

x = sonar_data.drop(columns = 60 , axis = 1)
y = sonar_data[60]
```

```python
print(x)
print(y)
```

Splitting into training and testing data

```python
x_train , x_test , y_train, y_test = train_test_split(x, y , test_size = 0.1,
                                              stratify= y , random_state = 1 )

# random_state -> Controls the shuffling of the dataset before splitting.
```

```python
print(x.shape , x_train.shape, x_test.shape)
```

```python
print(x_train)
print(y_train)
```

Model Training using Logisitic Regression

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regressionnmodel with training data
model.fit(x_train , y_train)
```

Model Evaluation

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction , y_train)
```

```python
print("Accuracy on training data : " , training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction , y_test)
```

```python
print("Accuracy on test data : " , test_data_accuracy)
```
